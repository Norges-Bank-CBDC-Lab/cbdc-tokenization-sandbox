# Script runner

## How to use Jupyter Notebooks

The script runner is based on Jupyter notebooks. Notebooks are structured into cells which contain either raw text, markdown-formatted text or Python code. Running a cell
will format markdown cells or run the Python code.

To run a cell, click the **Run** button or press either `Shift + Enter` (which takes your cursor to the next cell) or `Ctrl + Enter` (to stay on the current cell).

To add a cell, click on **Insert a cell below** or press `b` (for this, a cell needs to be selected, but you cannot be inside the cell, editing it; press `Esc` to exit the editing mode). To change the cell type to **Markdown**, use the dropdown menu at the top.

If you want to keep the changes you make to a notebook, make sure to download it.

## Sample notebooks
